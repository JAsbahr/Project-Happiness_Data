```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df_2015 = pd.read_csv("happiness_data_2015.csv")
df_2016 = pd.read_csv("happiness_data_2016.csv")
df_2017 = pd.read_csv("happiness_data_2017.csv")
df_2018 = pd.read_csv("happiness_data_2018.csv")
df_2019 = pd.read_csv("happiness_data_2019.csv")
```

Let's get all data into one DataFrame

```python
df_2015.head()
```

```python
year_2015 = np.repeat(2015, len(df_2015))
df_2015["Year"] = year_2015
df_2015.head()
```

```python
year_2019 = np.repeat(2019, len(df_2019))
df_2019["Year"] = year_2019
df_2019.head()
```

Problem: column names aren't the same and amount of kind of columns aren't the same as well. We we need to rename the columns to a standard and only take those columns that are listed in every year

```python
change = {"Happiness Rank":"Overall rank","Country":"Country or region", "Happiness Score":"Score", "Economy (GDP per Capita)":"GDP per capita","Family":"Social support","Health (Life Expectancy)":"Healthy life expectancy","Freedom":"Freedom to make life choices","Trust (Government Corruption)":"Perceptions of corruption"}
df_2015.rename(columns=change, inplace=True)
df_2015.head()
```

```python
df_2015.drop(columns=["Region", "Standard Error","Dystopia Residual"], inplace=True)
df_2015.head()
```

```python
df = df_2019.append(df_2015, ignore_index=True)
df.head()
```

```python
df.tail()
```

```python
len(df) == len(df_2015) + len(df_2019)
```

Same procedure for 2016, 2017 and 2018

```python
df = df.append(df_2016, ignore_index=True)
```

```python
df.tail()
```

```python
len(df) == len(df_2015) + len(df_2019) + len(df_2016)
```
