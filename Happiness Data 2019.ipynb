<h2>Analysing world happiness data 2019</h2>
<h3>Which factors lead to a better score?</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("happiness_data_2019.csv")
data.head()
```

Overview about factors

```python
data.hist(figsize=(8,8));
```

```python
pd.plotting.scatter_matrix(data, figsize=(15,15));
```

Overview about all correlations

```python
data.corr()
```

Correlations with "Score" column, first three are above 0.75

```python
data.corr().iloc[2:, 1]
```

Visual for correlations with "Score" column

```python
data.corr().iloc[2:, 1].plot(kind="bar")
```

Correlation between "GDP per capita" and "Score"

```python
af = plt.figure(figsize=(4,4))

axes = af.add_axes([0,0,1,1])
axes.plot(data["GDP per capita"], data["Score"], "b")
```

Correlation between GDP per capita, Social support, Healthy life expenctancy and Score as scatter plots

```python
data.plot(x="GDP per capita", y="Score", kind="scatter")
data.plot(x="Social support", y="Score", kind="scatter")
data.plot(x="Healthy life expectancy", y="Score", kind="scatter")
```

```python
data.to_csv("data.csv", index=False)
```
